```python
def remove(file_names):
    dir1 = "/home/mingi/data/vistext/data/test_images/vertical"
    dir2 = "/home/mingi/data/vistext/data/test_images/horizontal"
    for file_name in file_names:
        file_name = str(file_name) + ".png"
        os.remove(os.path.join(dir1, file_name))
        os.remove(os.path.join(dir2, file_name))
```

```python
dir = "/home/mingi/data/vistext/data/test_images/horizontal"
for i, file_name in enumerate(os.listdir(dir)):
    if i < 500:
        continue
    print(file_name)
    image = Image.open(os.path.join(dir, file_name)).convert("RGB")
    plt.imshow(image)
    plt.axis("off")
    plt.show()

    if i % 600 == 590:
        break
```
