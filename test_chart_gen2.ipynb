```python
import matplotlib.pyplot as plt
import numpy as np

# Example data for total gate revenues from sporting events from 2006 to 2015
years = np.arange(2006, 2016)
revenues = [14.5, 15.1, 15.4, 15.1, 15.2, 15, 15.6, 16.2, 16.8, 17.4]

# Create the plot with a dark background
plt.style.use('dark_background')

# Plotting the data
plt.fill_between(years, revenues, color='dodgerblue', alpha=0.7)

# Add labels and title
plt.xlabel('Year', fontsize=12)
plt.ylabel('North America', fontsize=12)
plt.title('Total gate revenues from sporting events worldwide by region from 2006 to 2015\n(in billion U.S. dollars)', fontsize=12, weight='bold')

# Set axis limits
plt.ylim(0, 20)

# Set x-ticks and y-ticks
plt.xticks(np.arange(2006, 2016, step=2))
plt.yticks(np.arange(0, 20, step=5))

# Remove plot borders (spines)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

# Display the plot
plt.tight_layout()
plt.savefig('/home/mingi/data/test_charts3/Total_gate_revenues.png')
plt.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data for total gate revenues from sporting events from 2006 to 2015
years = np.arange(2006, 2016)
revenues = [14.5, 15.1, 15.4, 15.1, 15.2, 15, 15.6, 16.2, 16.8, 17.4]

# Create the plot with a dark background and x and y axes swapped
plt.style.use('dark_background')

# Plotting the data with swapped axes
plt.fill_betweenx(years, revenues, color='dodgerblue', alpha=0.7)

# Add labels and title
plt.ylabel('Year')
plt.xlabel('Total gate revenues (in billion U.S. dollars)')
plt.title('Total gate revenues from sporting events worldwide by region from 2006 to 2015\n(in billion U.S. dollars)', fontsize=12, weight='bold')

# Set axis limits
plt.xlim(0, 20)

# Set y-ticks to be in steps of 2 years
plt.yticks(np.arange(2006, 2016, step=2))

# Remove plot borders (spines)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

# Display the plot
plt.tight_layout()
plt.savefig('/home/mingi/data/test_charts3/Total_gate_revenues_hor.png')
plt.close()
```

```python
import matplotlib.pyplot as plt

# Data
countries = ['Russia', 'Canada', 'USA', 'China', 'Brazil', 'Australia', 'India', 'Argentina', 'Kazakhstan']
areas = [17098242, 9984670, 9833517, 9596960, 8515770, 7741220, 3287263, 2780400, 2724900]

# Create a horizontal bar chart with blue color for all bars
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(countries, areas, color='#4682B4')  # '#4682B4' is a shade of blue

# Set background color to light gray
fig.patch.set_facecolor('#F5F5F5')  # Light gray background
ax.set_facecolor('#F5F5F5')

# Remove x-axis labels and ticks
ax.set_xticks([])

# Set country names (y-axis labels) to black
ax.tick_params(axis='y', colors='black', labelsize=12)

# Add value labels inside the bars with black text
for index, value in enumerate(areas):
    ax.text(value, index, f'{value:,}', va='center', ha='left', color='black', fontsize=10)

# Show the plot without additional padding
plt.tight_layout()
plt.savefig('/home/mingi/data/test_charts3/country.png')
plt.close()
```

```python
import matplotlib.pyplot as plt

# Data
countries = ['Russia', 'Canada', 'USA', 'China', 'Brazil', 'Australia', 'India', 'Argentina', 'Kazakhstan']
areas = [17098242, 9984670, 9833517, 9596960, 8515770, 7741220, 3287263, 2780400, 2724900]

# Create a vertical bar chart with blue color for all bars
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(countries, areas, color='#4682B4')  # '#4682B4' is a shade of blue

# Set background color to light gray
fig.patch.set_facecolor('#F5F5F5')  # Light gray background
ax.set_facecolor('#F5F5F5')

# Remove y-axis labels and ticks
ax.set_yticks([])

# Set country names (x-axis labels) to black
ax.tick_params(axis='x', colors='black', labelsize=12)

# Add value labels above the bars with black text
for index, value in enumerate(areas):
    ax.text(index, value, f'{value:,}', ha='center', va='bottom', color='black', fontsize=10)

# Show the plot without additional padding
plt.tight_layout()
plt.savefig('/home/mingi/data/test_charts3/country_hor.png')
plt.close()
```
